### Import Data

```python
import pandas as pd

# Etalab Datasets
etalab_df = pd.read_csv("etalab-datasets-2024-11-22-19-45.csv", delimiter=';', on_bad_lines='warn')
```

### Missing Value Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_missing_values(dataframe):
    plt.figure(figsize=(10, 6))
    sns.heatmap(dataframe.isnull(), cbar=False, cmap='viridis')
    plt.title("Missing Values Heatmap")
    plt.xlabel("Columns")
    plt.xticks(rotation=60)
    plt.ylabel("Rows")
    plt.show()

plot_missing_values(etalab_df)
```
